```python
#13) Using the user defined function convert the ‘rasied_amount_used’ column into a categorical column as follow-
#a) If the amount is less than 5 Millions then recode as TypeA.
#b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
#c) If the amount is greater than 7 Millions then recode as TypeC.
```

```python
def recode_amount_category(amount):
    if amount < 5000000:
        return 'TypeA'
    elif 5000000 <= amount < 7000000:
        return 'TypeB'
    elif amount > 7000000:
        return 'TypeC'
```

```python
master_dataframe['raised_amount_usd'].apply(recode_amount_category)
```
